```python
#Lektion 2, Exercise 1
total_amount = float(input("Total amount: "))
num_people = int(input("Number of people: "))

while True:
    tip_percentage = float(input("Tip percentage: "))
    if tip_percentage >= 10:
        break
    else:
        print("Tip percentage should be at least 10%.")

def calculate_split(total_amount, num_people, tip_percentage):

    return (total_amount * (1 + tip_percentage / 100)) / num_people

amount_per_person = calculate_split(total_amount, num_people, tip_percentage)
print(f"Each person owes: {amount_per_person:.2f}")
```

```python
#Lektion 2, Exercise 2

def check_password(password):
    special_characters = """ !"#$%&'()*+,-./:;<=>?@[\]^_`{|}~"""
    if (
        len(password) >= 8 and
        any(char.isdigit() for char in password) and
        any(char in special_characters for char in password)
        ):
        return True
    return False


while True:
    password = input("Enter your password: ")
    if check_password(password) == True:
        print("Password accepted.")
        break
    else:
        print("Password must be at least 8 characters long, and contain at least one number, and one special character.")




```

```python
#Lektion 2, Exercise 3

while True:

    weight = float(input("Enter your weight in kg: "))
    height = float(input("Enter your height in meters: "))

    bmi = weight / (height ** 2)

    def get_health_status(bmi):
        if bmi < 18.5:
            return "Underweight"
        elif 18.5 <= bmi < 25:
            return "Normal weight"
        elif 25 <= bmi < 30:
            return "Overweight"
        else:
            return "Obesity"

    health_status = get_health_status(bmi)
    print(f"Your BMI is {bmi:.2f}, which is considered {health_status}.")

    continue_prompt = input("Do you want to calculate another BMI? (1 = yes/ 0 = no): ")
    if continue_prompt == "0":
        break
```

```python
#Lektion 2, Exercise 4

gc_count = 0
dna_seq = input("Enter DNA sequence: ").strip().upper()

for char in dna_seq:
    if char not in "ACGT":
        print("Invalid DNA sequence. Only A, C, G, and T are allowed.")
        break

    if char in "GC":
        gc_count += 1

gc_content = (gc_count / len(dna_seq)) * 100
print(f"GC content: {gc_content:.2f}%")

def complementary_dna(dna_seq):
    complement = {'A': 'T', 'C': 'G', 'G': 'C', 'T': 'A'}


    return ''.join(complement[char] for char in dna_seq)

complement_seq = complementary_dna(dna_seq)
print(f"Complementary DNA sequence: {complement_seq}")
```
