```python
#Lektion 2, Exercise 1
total_amount = float(input("Total amount: "))       # Spørger brugeren om totale beløb for regningen, i float format
num_people = int(input("Number of people: "))       # Spørger brugeren om antal personer, i integer format, så der ikke kan gives halve mennesker

while True:                                             # Laver en while løkke, som bliver ved med at spørge om tips indtil en "fair" procent er givet
    tip_percentage = float(input("Tip percentage: "))
    if tip_percentage >= 10:                            # Den "fair" procent er sat til 10%, så hvis brugeren giver en procent under det, bliver de bedt om at give en højere procent
        break                                           # Hvis 10% eller mere er givet, brydes løkken
    else:
        print("Tip percentage should be at least 10%.")

def calculate_split(total_amount, num_people, tip_percentage):      # Funktion er tager totale pris, antal mennesker, og tip %, og returnerer hvor meget hver person skal betale

    return (total_amount * (1 + tip_percentage / 100)) / num_people   # Udregning

split = calculate_split(total_amount, num_people, tip_percentage)       # Funktionen kaldes
print(f"Each person owes: {split:.2f}")                             # Printer en formatted string 
```

Her starter vi med at spørge brugeren om input for det totale beløb, og antal personer. Så startes et while loop, så brugeren bliver ved med at blive promtet, indtil de inputter en "fair" tip procent. Derefter beregnes det totale beløb, inklusiv tip, og deles med antal personer inde i en funktion. Til sidst kaldes funktionen, og resultatet printes i en formatted string.

```python
#Lektion 2, Exercise 2

def check_password(password):                                       # Funktion der tjekker om password er stærkt nok
    special_characters = r""" $"%!&'()*+,-./:;<=>?@[\]^_`{|}~# """   # String der indeholder alle special characters, som tjekkes imod
    if (
        len(password) >= 8 and                              # Er password mindst 8 karakterer langt
        any(char.isdigit() for char in password) and            # Er der mindst et tal i password
        any(char in special_characters for char in password)    # Er der mindst et special character i password
        ):
        return True
    return False


while True:                                         # While løkke, så programmet bliver ved med at promte brugeren
    password = input("Enter your password: ")
    if check_password(password) == True:
        print("Password accepted.")
        break
    else:                                           # Hvis password ikke er stærkt nok, bliver der bedt om at prøve igen, og får at vide hvad kravene er
        print("Password must be at least 8 characters long, and contain at least one number, and one special character.")       
```

Der defineres en funktion som skal checke koden. Der laves en string som indeholder alle specielle karakterer, den skal være raw, så tegnene indeni ikke forvirrer/ødelægger koden. Så tjekkes der for om koden er lang nok, derefter om koden indeholder et tal, og til sidst om koden indeholder et tegn som matcher en af de tegn i special_characters stringen. 
Der startes så et while loop, så brugeren kan blive ved med at blive promptet. heri spørger den om kode inputtet, og checker om koden overholder kravene, hvis ikke, så printes kravene til brugeren, og man kan prøve igen.

```python
#Lektion 2, Exercise 3

while True:                 # While løkke, så programmet kan køre flere gange, hvis brugeren ønsker det

    weight = float(input("Enter your weight in kg: "))      # Bruger input
    height = float(input("Enter your height in meters: "))

    bmi = weight / (height ** 2)        # BMI udregnes

    def get_health_status(bmi):         # Funktion der tager BMI og returnerer den kategori BMI'en falder ind under
        if bmi < 18.5:
            return "Underweight"
        elif 18.5 <= bmi < 25:
            return "Normal weight"
        elif 25 <= bmi < 30:
            return "Overweight"
        else:
            return "Obesity"

    health_status = get_health_status(bmi)          #funktionen kaldes, og printes som en formatted string
    print(f"Your BMI is {bmi:.2f}, which is considered {health_status}.")

    continue_prompt = input("Do you want to calculate another BMI? (0 = no / any other key = yes): ")
    if continue_prompt == "0":          # Hvis brugeren ikke ønsker at fortsætte, indtast 0, og løkken breaker
        break
```

Der startes en while løkke så programmet kan køre flere gange. Der inputtes højde og vægt, hvor der fortælles brugeren hvilken enhed. Så udregnes BMI, og funktionen get_health_status kategoriserer hvilken klasse BMI'en er i. Så kaldes funktionen og resultatet printes i en formatted string. Til sidst spørges brugeren om de vil køre programmet igen, eller stoppe det. 0 indtastes for at stoppe det, alle andre taster, for at køre endnu en gang.

```python
#Lektion 2, Exercise 4

gc_count = 0            # Initialiserer gc_count som 0
dna_seq = input("Enter DNA sequence: ").strip().upper()         # Den inputtede DNA sekvens strippes for mellemrum, og gøres uppercase

for char in dna_seq:                # checker hvert tegn i den inputtede string
    if char not in "ACGT":          # Hvis der er tegn i stringen som ikker indgår i DNA normalt, break og meld fejl
        print("Invalid DNA sequence. Only A, C, G, and T are allowed.")
        break

    if char in "GC":        # Hvis tegnet er G eller C, tælles det i gc_count
        gc_count += 1

gc_content = (gc_count / len(dna_seq)) * 100        # beregner procentdelen af GC i den inputtede DNA sekvens
print(f"GC content: {gc_content:.2f}%")             # print som en formatted string

def complementary_dna(dna_seq):                 # Funktion der tager en DNA sekvens, og returnerer den komplementære sekvens
    complement = {'A': 'T', 'C': 'G', 'G': 'C', 'T': 'A'}       # Dict der forklarer hvad det komplementære basepar er for hver base

    return ''.join(complement[char] for char in dna_seq)        

complement_seq = complementary_dna(dna_seq)
print(f"Complementary DNA sequence: {complement_seq}")
```

Der sker et par ting inde i complementary_dna funktionen. Den tager en DNA sekvens string som parameter, og det første der sker er at en dict hvor der defineres hvilken base der er den komplementære for hver base. Så keysne er baserne, og valuesne er deres komplementære base. Dette bruges så nedenunder, her ietererer vi igennem hvert tegn i dna_seq, og finder valuen(den komplementære) for den key som vi er ved i dna_seq. Alle disse komplementære værdier joines så i en enkelt string, og vi har resultatet.
