# Exercise 1

Her starter vi med at spørge brugeren om input for det totale beløb, og antal personer. Så startes et while loop, så brugeren bliver ved med at blive promtet, indtil de inputter en "fair" tip procent. Derefter beregnes det totale beløb, inklusiv tip, og deles med antal personer inde i en funktion. Til sidst kaldes funktionen, og resultatet printes i en formatted string.

```python
total_amount = float(input("Total amount: "))       # Spørger brugeren om totale beløb for regningen, i float format
num_people = int(input("Number of people: "))       # Spørger brugeren om antal personer, i integer format, så der ikke kan gives halve mennesker

while True:                                             # Laver en while løkke, som bliver ved med at spørge om tips indtil en "fair" procent er givet
    tip_percentage = float(input("Tip percentage: "))
    if tip_percentage >= 10:                            # Den "fair" procent er sat til 10%, så hvis brugeren giver en procent under det, bliver de bedt om at give en højere procent
        break                                           # Hvis 10% eller mere er givet, brydes løkken
    else:
        print("Tip percentage should be at least 10%.")

def calculate_split(total_amount, num_people, tip_percentage):      # Funktion er tager totale pris, antal mennesker, og tip %, og returnerer hvor meget hver person skal betale

    return (total_amount * (1 + tip_percentage / 100)) / num_people   # Udregning

split = calculate_split(total_amount, num_people, tip_percentage)       # Funktionen kaldes
print(f"Each person owes: {split:.2f}")                             # Printer en formatted string 
```

# Exercise 2

Der defineres en funktion som skal checke koden. Der laves en string som indeholder alle specielle karakterer, den skal være raw, så tegnene indeni ikke forvirrer/ødelægger koden. Så tjekkes der for om koden er lang nok, derefter om koden indeholder et tal, og til sidst om koden indeholder et tegn som matcher en af de tegn i special_characters stringen. 
Der startes så et while loop, så brugeren kan blive ved med at blive promptet. heri spørger den om kode inputtet, og checker om koden overholder kravene, hvis ikke, så printes kravene til brugeren, og man kan prøve igen.

```python
def check_password(password):                                       # Funktion der tjekker om password er stærkt nok
    special_characters = r""" $"%!&'()*+,-./:;<=>?@[\]^_`{|}~# """   # String der indeholder alle special characters, som tjekkes imod
    if (
        len(password) >= 8 and                              # Er password mindst 8 karakterer langt
        any(char.isdigit() for char in password) and            # Er der mindst et tal i password
        any(char in special_characters for char in password)    # Er der mindst et special character i password
        ):
        return True
    return False


while True:                                         # While løkke, så programmet bliver ved med at promte brugeren
    password = input("Enter your password: ")
    if check_password(password) == True:
        print("Password accepted.")
        break
    else:                                           # Hvis password ikke er stærkt nok, bliver der bedt om at prøve igen, og får at vide hvad kravene er
        print("Password must be at least 8 characters long, and contain at least one number, and one special character.")       
```

# Exercise 3

Der startes en while løkke så programmet kan køre flere gange. Der inputtes højde og vægt, hvor der fortælles brugeren hvilken enhed. Så udregnes BMI, og funktionen get_health_status kategoriserer hvilken klasse BMI'en er i. Så kaldes funktionen og resultatet printes i en formatted string. Til sidst spørges brugeren om de vil køre programmet igen, eller stoppe det. 0 indtastes for at stoppe det, alle andre taster, for at køre endnu en gang.

```python
while True:                 # While løkke, så programmet kan køre flere gange, hvis brugeren ønsker det

    weight = float(input("Enter your weight in kg: "))      # Bruger input
    height = float(input("Enter your height in meters: "))

    bmi = weight / (height ** 2)        # BMI udregnes

    def get_health_status(bmi):         # Funktion der tager BMI og returnerer den kategori BMI'en falder ind under
        if bmi < 18.5:
            return "Underweight"
        elif 18.5 <= bmi < 25:
            return "Normal weight"
        elif 25 <= bmi < 30:
            return "Overweight"
        else:
            return "Obesity"

    health_status = get_health_status(bmi)          #funktionen kaldes, og printes som en formatted string
    print(f"Your BMI is {bmi:.2f}, which is considered {health_status}.")

    continue_prompt = input("Do you want to calculate another BMI? (0 = no / any other key = yes): ")
    if continue_prompt == "0":          # Hvis brugeren ikke ønsker at fortsætte, indtast 0, og løkken breaker
        break
```

# Exercise 4

Der sker et par ting inde i complementary_dna funktionen. Den tager en DNA sekvens string som parameter, og det første der sker er at en dict hvor der defineres hvilken base der er den komplementære for hver base. Så keysne er baserne, og valuesne er deres komplementære base. Dette bruges så nedenunder, her ietererer vi igennem hvert tegn i dna_seq, og finder valuen(den komplementære) for den key som vi er ved i dna_seq. Alle disse komplementære værdier joines så i en enkelt string, og vi har resultatet.

```python
gc_count = 0            # Initialiserer gc_count som 0
dna_seq = input("Enter DNA sequence: ").strip().upper()         # Den inputtede DNA sekvens strippes for mellemrum, og gøres uppercase

for char in dna_seq:                # checker hvert tegn i den inputtede string
    if char not in "ACGT":          # Hvis der er tegn i stringen som ikker indgår i DNA normalt, break og meld fejl
        print("Invalid DNA sequence. Only A, C, G, and T are allowed.")
        break

    if char in "GC":        # Hvis tegnet er G eller C, tælles det i gc_count
        gc_count += 1

gc_content = (gc_count / len(dna_seq)) * 100        # beregner procentdelen af GC i den inputtede DNA sekvens
print(f"GC content: {gc_content:.2f}%")             # print som en formatted string

def complementary_dna(dna_seq):                 # Funktion der tager en DNA sekvens, og returnerer den komplementære sekvens
    complement = {'A': 'T', 'C': 'G', 'G': 'C', 'T': 'A'}       # Dict der forklarer hvad det komplementære basepar er for hver base

    return ''.join(complement[char] for char in dna_seq)        

complement_seq = complementary_dna(dna_seq)
print(f"Complementary DNA sequence: {complement_seq}")
```

# Dosimetry Assignment
### Lav en Dosimeter klasse der kan loade, holde og analysere dosimetri data. Sammenlign dosimetri data for virksomhederne

Nedenunder har vi lavet klassen Dosemetry, der under oprettelse gives ét Person ID, fra én person.

Metoden load_patient_log() gør det muligt at indlæse og formatere Dosimetri data ud fra en .txt fil.

Metoden load_fyraften() indlæser data fra Fyraften.csv filen og formaterer disse.

limit_patient_data_by_fyraften() fjerner alle data efter fyraften for person.

calculate_total_leq() udregner Total L_eq og calculate_lex8() udregner L_ex,8.

Til sidst samler run_full_analysis() ovenstående metoder og benytter disse til at udregne L_eq, L_ex,8 og total arbejdstid for én person.

```python
import pandas as pd
import numpy as np

# Klassen modtager én patient_id og behandler data på tværs a filer for denne.
class Dosemetry:
    def __init__(self, patient_id):
        self.patient_id = str(patient_id)
        self.patient_df = pd.DataFrame()
        self.firma = None
        self.fyraften_df = pd.DataFrame()
        self.filtered_df = pd.DataFrame()
        self.total_hours_worked = None
        self.Total_Leq = None
        self.Lex8 = None

    #Indlæser og konverterer dosimetri data fra PATIENT_ID.txt fil til Pandas DataFrame
    def load_patient_log(self):
        file_path = f"DosimetriData/{self.patient_id}.txt"

        #Skip øverste 11 linjer og sorter data ud fra mellemrum i teksten.
        df = pd.read_csv(file_path, sep=r"\s+", header=None, skiprows=11, engine="python")
        df.columns = ['Datapoint', 'Flags', 'Date', 'Time', 'Leq', 'CPEAK', 'Extra']

        #Kombiner Dato og Tid til én værdi og gem denne
        df['Datetime'] = pd.to_datetime(df['Date'] + ' ' + df['Time'], format='%d.%m.%Y %H:%M:%S', errors='coerce')
        # Formater Leq decimal fra , til .
        df['Leq'] = pd.to_numeric(df['Leq'].str.replace(',', '.', regex=False), errors='coerce')
        # Fjern rækker hvor Datetime eller Leq ikke eksisterer
        df = df.dropna(subset=['Datetime', 'Leq'])

        self.patient_df = df
        return df

    #Indlæser og konverterer data fra Fyraften.csv til Pandas Dataframe
    def load_fyraften(self):
        df = pd.read_csv("DosimetriData/Fyraften.csv", sep=';')
        df.columns = [col.strip().lower() for col in df.columns]

        if 'patientid' not in df.columns:
            if 'personid' in df.columns:
                df = df.rename(columns={'personid': 'patientid'})

        self.fyraften_df = df
        return df

    def get_firma(self):
        df = pd.read_csv("DosimetriData/Noeglefil.csv", sep=';')
        df.columns = [col.strip().lower() for col in df.columns]
        result = df.loc[df["personid"] == int(self.patient_id), "virksomhedsnavn"]
        if not result.empty:
            self.firma = result.iloc[0]
        else:
            self.firma = None

    # Fjerner alle målinger efter fyraften
    def limit_patient_data_by_fyraften(self):
        if self.fyraften_df.empty or self.patient_df.empty:
            self.filtered_df = self.patient_df
            self.total_hours_worked = None
            return self.patient_df

        row = self.fyraften_df[self.fyraften_df['patientid'].astype(str) == self.patient_id]

        if row.empty:
            self.filtered_df = self.patient_df
            self.total_hours_worked = None
            return self.patient_df

        fyraften_str = str(row.iloc[0]['fyraften'])

        if not self.patient_df.empty:
            first_date = self.patient_df.iloc[0]['Date']
            fyraften_datetime = pd.to_datetime(f"{first_date} {fyraften_str}", format='%d.%m.%Y %H.%M', errors='coerce')
        else:
            fyraften_datetime = pd.NaT

        filtered_df = self.patient_df[self.patient_df['Datetime'] < fyraften_datetime]
        self.filtered_df = filtered_df

        if not filtered_df.empty:
            start_time = filtered_df.iloc[0]['Datetime']
            end_time = filtered_df.iloc[-1]['Datetime']
            self.total_hours_worked = (end_time - start_time).total_seconds() / 3600.0
        else:
            self.total_hours_worked = None

        return filtered_df

    # Udregn total L_eq
    def calculate_total_leq(self):
        df = self.filtered_df

        if df.empty or 'Leq' not in df:
            return None

        leq_values = df['Leq'].dropna().values
        if len(leq_values) == 0:
            return None

        total_leq = 10 * np.log10(np.mean(10 ** (leq_values / 10)))
        return round(total_leq, 3)

    # Udregn L_ex,8
    def calculate_lex8(self):
        leq_t = self.calculate_total_leq()
        T = self.total_hours_worked

        if leq_t is None or T is None or T <= 0:
            return None

        lex8 = leq_t + 10 * np.log10(T / 8)
        return round(lex8, 3)

    def run_full_analysis(self):
        self.load_patient_log()
        self.load_fyraften()
        self.limit_patient_data_by_fyraften()

        self.Total_Leq = self.calculate_total_leq()
        self.Lex8 = self.calculate_lex8()

        if self.total_hours_worked is not None:
            self.total_hours_worked = round(self.total_hours_worked, 3)
```

Nedenunder vises eksempel hvor Total Leq, Lex,8 og total arbejdstid udregnes for én Person ID (129).

```python
P_129 = Dosemetry(129)
P_129.run_full_analysis()
total_leq = P_129.Total_Leq
lex8 = P_129.Lex8
hours = P_129.total_hours_worked
print(f"Total Leq: {total_leq:.3f}")
print(f"Lex,8: {lex8:.3f}")
print(f"Total hours worked: {hours:.3f}")
```

Vi kan nu udregne værdierne for alle PersonID, ved at samle disse i en liste og så loope igennem dem.

```python
# Script to run analysis for all patients
patient_ids = [129,131,133,142,188,228,242,254,268,386,444,453,480,495,501,537,541,681,689,690,693,711,712,725,728,735,769,779,782,789,815,820,841,864,924,940,955,961,973]
results = {}
for pid in patient_ids:
    patient = Dosemetry(pid)
    patient.run_full_analysis()
    patient.get_firma()
    results[pid] = {
        'total_leq': patient.Total_Leq,
        'lex8': patient.Lex8,
        'total_hours_worked': patient.total_hours_worked,
        'firma': patient.firma,
    }
# Print results for all patients
for pid, res in results.items():
    print(f"Patient {pid}: Total Leq = {res['total_leq']:.3f}, Lex,8 = {res['lex8']:.3f}, Hours Worked = {res['total_hours_worked']:.3f}, Firma = {res['firma']}")
```

Ud fra results dictionary kan vi aggregere l_ex,8 værdierne for hver firma og få middelværdien fra disse.
Pandas har en smart indbygget metode, groupby, som gør dette meget let.

Som forventet ser vi de højeste værdier hos Haard Metal (som man må forventer laver en del støj).

```python
oversigt_tabel = pd.DataFrame(results).T
oversigt_tabel_middel = oversigt_tabel.groupby("firma")["lex8"].agg(["mean"])
print(oversigt_tabel_middel)
```
